```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
def factory_soup(url):
    r = requests.get(url)
    data = r.text
    return BeautifulSoup(data, 'html.parser')
```

```python
soup_metadata = factory_soup('http://www.ies.ufpb.br/ojs2/index.php/ies/rt/metadata/343')
content = soup_metadata.find('div', id='content')
rows = content.table.find_all('tr', valign='top')
```

```python
metadata_document = []

for row in rows[1:]:
    elements = row.find_all('td')
    
    metadata = {
        'id': elements[0].get_text(),
        'description': elements[1].get_text(),
        'pkp': elements[2].get_text(),
        'document': elements[3].get_text(),
    }
    
    metadata_document.append(metadata)
```

```python
def filter_field(id, description, pkp):
    return [d['document'].strip() for d in metadata_document if d['id'] in id and d['description'] in description and d['pkp'] in pkp]
    
title = filter_field('1.', 'Título', 'Título do documento')[0]
authors = filter_field('2.', 'Autor', 'Nome do autor, afiliação institucional, país')
keywords = filter_field('3.', 'Assunto', 'Palavras-chave(s)')[0]
summary = filter_field('4.', 'Descrição', 'Resumo')[0]
url_view = filter_field('10.', 'Identificador', 'Identificador de Recurso Uniforme (URI)')[0]
soup_url_pdf = factory_soup(url_view)
url_pdf = soup_url_pdf.find('div', id='articleFullText').a['href']

fields = {
    'title': title, 
    'authors': authors,
    'keywords': keywords, 
    'summary': summary, 
    'url_view': url_view, 
    'url_pdf': url_pdf,
    
    'raw_data': metadata_document,
    'html_doc': soup_metadata,
}
print(fields['html_doc'])
```
